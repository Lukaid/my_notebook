## 페이지 넘기기 추가

```python
import requests
from bs4 import BeautifulSoup
import re

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.190 Safari/537.36"}
```

```python
for i in range(1, 6): # 1 ~ 5 페이지의 정보 가져오기
    print()
    print("페이지 :", i)
    print()

    url = "https://www.coupang.com/np/search?q=%EB%85%B8%ED%8A%B8%EB%B6%81&channel=user&component=&eventCategory=SRP&trcid=&traid=&sorter=scoreDesc&minPrice=&maxPrice=&priceRange=&filterType=&listSize=36&filter=&isPriceRange=false&brand=&offerCondition=&rating=0&page={}&rocketAll=false&searchIndexingToken=&backgroundColor=".format(i)

    res = requests.get(url, headers = headers)
    res.raise_for_status() # User-Agent값을 지정해주니까 서버에서 안막는 듯
    soup = BeautifulSoup(res.text, "lxml")

    items = soup.find_all('li', attrs={'class':re.compile('^search-product')})

    for item in items:

        ad_badge = item.find("span", attrs={'class':'ad-badge-text'})
        if ad_badge:
            #print("<< 광고 상품 제외 >>")
            continue

        name = item.find("div", attrs={'class':'name'}).get_text() # 제품명
        price = item.find("strong", attrs={'class':'price-value'}).get_text() # 가격
        rate = item.find("em", attrs={'class':'rating'}) # 평점
        rate_count = item.find("span", attrs={'class':'rating-total-count'}) # 리뷰수

        if "Apple" in name:
            #print("<< Apple 상품 제외 >>")
            continue

        if rate:
            rate = rate.get_text()
            if float(rate) >= 4.5:
                pass
            else:
                #print("<< 평점 4.5 미만 제외 >>")
                continue
        else:
            rate = '평점없음'
            #print("<< 평점 없는 상품 제외 >>")
            continue

        if rate_count:
            rate_count = rate_count.get_text()
            if float(rate_count[1:-1]) >= 100:
                pass
            else:
                #print("<< 리뷰 100개 미만 제외 >>")
                continue
        else:
            rate_count = '리뷰 없음'
            #print("<< 리뷰 없는 상품 제외 >>")
            continue
        
        link = item.find("a", attrs={"class":"search-product-link"})["href"]
        #print(name, "||", price,"||", rate,"||", rate_count)
        print(f"제품명 : {name}")
        print(f"가격: {price}")
        print(f"평점 : {rate}점 {rate_count[1:-1]}개")
        print(f"바로가기 : https://www.coupang.com/{link}")
        print('--------------------------------------------------------------------------------------------------------------')
        print()
```
