```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams["font.size"] = 5
plt.rcParams["font.family"] = 'NanumGothic'
plt.rcParams['xtick.labelsize'] = 8 
plt.rcParams['ytick.labelsize'] = 8

건설 = pd.read_csv("건설_0208.csv", encoding = 'CP949')

name = list(건설['시명'])
baechul = list(건설['2020 건설폐기물 배출량'])
churi = list(건설['2020 건설폐기물 처리량'])

fig, ax = plt.subplots()
ax.scatter(baechul, churi, s = 1)

plt.title('2020년 건설폐기물', fontsize = 15)
plt.xlabel('지역 내 건설폐기물 배출량 비중', fontsize = 10)
plt.ylabel('지역 내 건설폐기물 처리량 비중', fontsize = 10)

for i, txt in enumerate(name):
    ax.annotate(txt, (baechul[i], churi[i]), rotation=45)

plt.savefig('./2020_건설.png', dpi = 400)
```
