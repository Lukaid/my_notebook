# How to Install


## Using Make

```bash
$ git clone https://github.com/facebookresearch/fastText.git
$ cd fastText
$ mkdir build && cd build && cmake ..

```

```bash
$ git clone https://github.com/facebookresearch/fastText.git
```

````python
from gensim.models import FastText
import pandas as pd
````

````python
result = pd.read_table('review.sorted.uniq.refined.tsv.text.tok', encoding='utf-8', header = None)
````

````python
result
````

````python
from konlpy.tag import Mecab

# 형태소분석 - kkma 명사 
mecab = Mecab()

line_list = []

for line in result[0]:
    line = mecab.nouns(line)
    line_list.append(line)

print("- 불러온 문서 :", len(line_list), "문장")
````

````python
word_frequency = {}
noun_list = []
# stop_list는 위에서 정의
stop_list = []
line_number = 0
for line in line_list[:]:
    line_number += 1
    print(str(line_number) + "/" + str(len(line_list)), end="\r")
    noun = []
    for word in line:
        if word.split("/")[0] not in stop_list and len(word.split("/")[0]) > 1:  
            noun.append(word.split("/")[0])
            if word not in word_frequency.keys():
                word_frequency[word] = 1
            else:
                word_frequency[word] += 1
    noun_list.extend(noun)
````

````python
noun_doc = ' '.join(noun_list)
noun_doc = noun_doc.strip()
````

````python
noun_doc
````

````python
model = FastText(noun_doc, size=100, window=5, min_count=5, workers=4, sg=1)
````

````python
model.wv.most_similar('배송')
````

````python
import nltk
nltk.download('punkt')
import urllib.request
import zipfile
from lxml import etree
import re
from nltk.tokenize import word_tokenize, sent_tokenize
````

````python
mecab = Mecab()
tokenized_data = []
stopwords = []
for sentence in result[0]:
    temp_X = mecab.morphs(sentence) # 토큰화
    temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
    tokenized_data.append(temp_X)
````

````python
# 리뷰 길이 분포 확인
print('리뷰의 최대 길이 :',max(len(l) for l in tokenized_data))
print('리뷰의 평균 길이 :',sum(map(len, tokenized_data))/len(tokenized_data))
# plt.hist([len(s) for s in tokenized_data], bins=50)
# plt.xlabel('length of samples')
# plt.ylabel('number of samples')
# plt.show()
````

여기서 Word2Vec의 하이퍼파라미터값은 다음과 같습니다.
- size = 워드 벡터의 특징 값. 즉, 임베딩 된 벡터의 차원.
- window = 컨텍스트 윈도우 크기
- min_count = 단어 최소 빈도 수 제한 (빈도가 적은 단어들은 학습하지 않는다.)
- workers = 학습을 위한 프로세스 수
- sg = 0은 CBOW, 1은 Skip-gram.

````python
from gensim.models import Word2Vec
from gensim.models import FastText

model_w = Word2Vec(sentences=line_list, size = 100, window = 5, min_count = 5, workers = 4, sg = 0)
model_f = FastText(sentences=line_list, size = 100, window = 5, min_count = 5, workers = 4, sg = 0)
````
